```python
# 3d blobs
blobs, y = datasets.make_blobs(n_samples=n_samples, n_features=3, random_state=8)
print(y)
transformation = [[0.2, -0.3, 5], [0.6, -0.6, 1], [-0.4, 0.8, 3]]
blobs = np.dot(blobs, transformation)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(blobs[:, 0], blobs[:, 1], blobs[:, 2], 'ko')
```

```python
plt.plot(blobs[:, 0], blobs[:, 1], 'k.')
```

```python
plt.plot(blobs[:, 0], blobs[:, 2], 'k.')
```

```python
plt.plot(blobs[:, 1], blobs[:, 2], 'k.')
```

```python
# components  > the number of new dims
pca = PCA(n_components=2)
X = pca.fit_transform(roll)

# 2 new features with meaning keep the info of main data > none of the above
plt.plot(X[:, 0], X[:, 1], 'b.')
```

```python
pca = PCA(n_components=2)
X = pca.fit_transform(s_curve)

plt.plot(X[:, 0], X[:, 1], 'b.')
```

```python
pca = PCA(n_components=2)
X = pca.fit_transform(blobs)

plt.plot(X[:, 0], X[:, 1], 'b.')
```
